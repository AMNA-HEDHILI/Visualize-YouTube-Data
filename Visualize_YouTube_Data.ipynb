```python
from google.colab import files

uploaded = files.upload()
```

```python
import io
import pandas as pd
df=pd.read_csv(io.BytesIO(uploaded['most_subscribed_youtube_channels.csv']))
display(df)
```

```python
df = df.replace(',', '')
display(df)
```

```python
import plotly.express as px

fig = px.bar(
    df,
    x="category",
    y="subscribers",
    title="Total Subscribers by Category",
    labels={"subscribers": "Total Subscribers"},
    text_auto=True
)

fig.show()
```

```python
df = df.dropna(subset=["category"]).copy()

df.loc[:, "subscribers"] = (
    df["subscribers"]
      .astype(str)
      .str.replace(",", "", regex=True)
      .astype(float)
)

```

```python
import plotly.express as px

fig = px.pie(
    df,
    values="subscribers",
    names="category",
    title="Subscriber Share by YouTube Category"
)

fig.update_traces(textinfo="percent+label")
fig.show()



```

```python
import plotly.express as px

fig = px.box(
    df,
    y="started",
    title="Distribution of YouTube Channel Start Years"
)

fig.show()
```

```python
num_cols = ["subscribers", "video views", "video count"]

df[num_cols] = (
    df[num_cols]
        .astype(str)
        .replace("0", None)
        .apply(lambda col: col.str.replace(",", "", regex=True))
        .astype(float)
)

```

```python
df = df.dropna(subset=["subscribers", "video count"])
```

```python
fig = px.scatter(
    df,
    x="video count",
    y="subscribers",
    color="category",
    size="video views",
    hover_name="Youtuber",
    title="Subscribers vs Video Count (Bubble Size = Total Views)",
    log_x=True,
    log_y=True
)

fig.show()
```
